### Problem 2.1
The cart pendulum equations-of-motion are:

\begin{eqnarray}
\ddot y + \frac{mg}{M} \theta &=& \frac{f}{M} \\
\ddot \theta - \frac{M+m}{M \ell} g \theta &=& -\frac{f}{M \ell}
\end{eqnarray}

$\tau = rf$, $r$ is the torque-to-linear-force factor.  For the electric motor:
\begin{eqnarray}
\tau &=& -\frac{k^2}{R} \omega + \frac{K}{R}e \\
&=& rf \\
\implies f &=& -\frac{k^2}{Rr} \omega + \frac{K}{Rr}e
\end{eqnarray}

The linear acceleration, $\ddot y$, is equal to $\ddot y = r \dot \omega \implies \dot y = r \omega$ by Newton's law.  This implies:
\begin{eqnarray}
f &=& -\frac{k^2}{Rr^2} \dot y + \frac{K}{Rr}e
\end{eqnarray}

Plug in $f$ to the first equation:
\begin{eqnarray}
\ddot y + \frac{mg}{M} \theta &=& -\frac{k^2}{MRr^2} \dot y + \frac{K}{MRr}e \\
\ddot \theta - \frac{M+m}{M \ell} g \theta &=& \frac{k^2}{MRr^2 \ell} \dot y - \frac{K}{MRr \ell}e
\end{eqnarray}

Move state derivatives to the LHS:
\begin{eqnarray}
\ddot y +\frac{k^2}{MRr^2} \dot y + \frac{mg}{M} \theta &=& \frac{K}{MRr}e \\
\ddot \theta - \frac{k^2}{MRr^2 \ell} \dot y - \frac{M+m}{M \ell} g \theta  &=& - \frac{K}{MRr \ell}e
\end{eqnarray}

### Problem 2.2
The Lagrange equations when the angular component is being driven by input:
\begin{eqnarray}
\frac{d}{dt} \biggl ( \frac{\partial L}{\partial \dot y} \biggr ) - \frac{\partial L}{\partial y} &=& 0 \\
\frac{d}{dt} \biggl ( \frac{\partial L}{\partial \dot \theta} \biggr ) - \frac{\partial L}{\partial \theta} &=& u
\end{eqnarray}

This leads to the two EOM:
\begin{eqnarray}
(M+m) \ddot y + m \ell \ddot \theta &=& 0 \\
m \ddot y + m \ell \ddot \theta - mg \theta &=& u
\end{eqnarray}

Reduce into two second-order equations for $y$ and $\theta$:
\begin{eqnarray}
\ddot y &=& -\frac{u}{M} - \frac{m}{M} g \theta \\
\ddot \theta &=& \frac{M+m}{M \ell} \bigl ( \frac{u}{m} + g \theta \bigr )
\end{eqnarray}

The input $u$ comes from motor torque:
\begin{eqnarray}
\tau &=& J \dot \omega \\ 
&=& -\frac{k^2}{R} \omega + ke \\
&=& -\frac{k^2}{R} \dot \theta + ke
\end{eqnarray}

The final driven equations are:
\begin{eqnarray}
\ddot y + \frac{k^2}{MR} \dot \theta + \frac{m}{M} g \theta &=& -\frac{k}{M} e \\
\ddot \theta + \frac{(M+m)k^2}{MmR \ell} \dot \theta - \frac{M+m}{M \ell} g \theta &=& \frac{M+m}{Mm \ell} k e
\end{eqnarray}

The second part of the question asks about the limit as $M \to \infty$.  When, $M \to \infty$, $\ddot y \to 0$ and $m \ell \ddot \theta \to -\frac{k^2}{R} \dot \theta + k e + mg \theta$, which is the EOM of a driven pendulum fixed at its pivot.

\begin{eqnarray}
\dot v_1 + \frac{3}{RC} v_1 - \frac{1}{RC} v_2 &=& \frac{2}{RC} e_0 \\

\end{eqnarray}

The reservoir term, $e_R$, is an exogenous input.

### Problem 2.4

The Lagrangian is
\begin{eqnarray}
L &=& T-V \\
&=& \frac{1}{2}M(\dot z_1^2 + \dot z_2^2) - \frac{1}{2}k(z_2 - z_1)^2
\end{eqnarray}

The variational equations for the unforced system are:
\begin{eqnarray}
\frac{d}{dt} \biggl ( \frac{\partial L}{\partial \dot z_1} \biggr ) - \frac{\partial L}{\partial z_1} = 0 &\implies& M \ddot z_1 - k(z_2 - z_1) &=& 0 \\
\frac{d}{dt} \biggl ( \frac{\partial L}{\partial \dot z_2} \biggr ) - \frac{\partial L}{\partial z_2} = 0 &\implies& M \ddot z_2 + k(z_2 - z_1) &=& 0
\end{eqnarray}

### Problem 2.6

#### a)
\begin{eqnarray}
\dot r &=& -\frac{\mathbf{r} \cdot \mathbf{v}}{r} \\
&=& -\frac{r V \cos \sigma}{r} \\
&=& -V \cos \sigma \\
\dot \lambda &=& -\frac{\mathbf{r} \times \mathbf{v}}{r^2} \\
&=& \frac{r V \sin \sigma}{r^2} \\
&=& \frac{V \sin \sigma}{r} \\
\dot \sigma &=& \dot \lambda + \dot \gamma \\
&=& \frac{V \sin \sigma}{r} + \frac{\mathbf{v} \times \mathbf{a}}{V^2} \\
&=& \frac{V \sin \sigma}{r} + \frac{V a}{V^2} \\
&=& \frac{V \sin \sigma}{r} + \frac{a}{V}
\end{eqnarray}
$\mathbf{v} \perp \mathbf{a}$ follows from the assumption that the commanded acceleration is normal to the velocity vector.

#### b)
The distance of closest approach, $z$, is the leg of a right triangle with hypotenuse $r$ and angle $\sigma$.  Therefore, $z = r \sin \sigma$.  Differentiating $z$ w.r.t time gives
\begin{eqnarray}
\dot z = \frac{ra \cos \sigma}{V}
\end{eqnarray}

#### c)
Using the relation $z = r \sin \sigma$, the differential equation for $\dot \lambda$ can be rewritten as:
\begin{eqnarray}
\dot \lambda &=& \frac{V \sin \sigma}{r} \\
&=& \frac{V r \sin \sigma}{r^2} \\
&\equiv& \frac{V z}{r^2}
\end{eqnarray}

Now, use the approximation $r/V \approx \bar T$ to get:
\begin{eqnarray}
\dot \lambda &=& \frac{V z}{r^2} \\
&=& \frac{1}{V} \biggl ( \frac{V}{r} \biggr )^2 z \\
&\approx& \frac{1}{V {\bar T}^2} z
\end{eqnarray}

For $z$:
\begin{eqnarray}
\dot z &=& \frac{r\cos \sigma}{V} a \\
&\approx& \frac{r}{V} a \\
&=& \bar T a, 
\end{eqnarray}
because $\cos \sigma \approx 1$ for $\sigma$ small.

The final transformed, approximate, differential equations are:
\begin{eqnarray}
\dot \lambda &\approx& \frac{1}{V {\bar T}^2} z \\
\dot z &\approx& \bar T a, 
\end{eqnarray}
